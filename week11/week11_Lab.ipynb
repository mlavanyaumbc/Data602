```python
from zipfile import ZipFile
from io import BytesIO
import urllib.request as urllib2
import pandas as pd

def get_data():
    r = urllib2.urlopen("https://github.com/msaricaumbc/DS_data/blob/master/ds602/people_wiki.zip?raw=true").read()
    file = ZipFile(BytesIO(r))
    people_wiki_csv = file.open("people_wiki.csv")
    people = pd.read_csv(people_wiki_csv)
    people.drop('URI', inplace=True, axis=1)
    return people

people = get_data()
people.head()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans

# Perform text vectorization
vectorizer = TfidfVectorizer(stop_words='english')
X = vectorizer.fit_transform(people['text'])

# Perform K-Means clustering
k = 6
kmeans = KMeans(n_clusters = k, init='random', n_init='auto')
clusters = kmeans.fit_predict(X)

# Print the cluster labels
for i in range(k):
    print(f"Cluster {i+1}:")
    for j, name in enumerate(people[clusters==i]['name']):
        print(f"{j+1}. {name}")
    print("")
```

## Question 5. How does the distribution of `salesprice` vary by make for cars made between 2010-2015?
- Use the 10 most common makes (your cleaned version of this column).  
- Create a graph that answers this question.

````python
#used the clean version of MAKE column and used the 10 most common makes.
meanNflPivot = df[df['year'] > 2009].pivot_table(index='MAKE', columns='year', values='sellingprice', aggfunc='count')
meanNflPivot.dropna(inplace=True)
meanNflPivot['count'] = meanNflPivot.sum(axis=1)
meanNflPivot.sort_values('count', ascending=False, inplace=True)
meanNflPivot[:10]
````

````python
import seaborn as sns

plt.figure(figsize=(9, 9))
sns.heatmap(meanNflPivot[:10], annot=True, fmt='.1f')
plt.xlabel('')
plt.ylabel('')
plt.title('distribution')
plt.show()
````
